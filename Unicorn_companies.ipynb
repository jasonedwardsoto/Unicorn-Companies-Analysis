```python
#Import libaries
import pandas as pd
from pathlib import Path
import numpy as py
import plotly.express as px
import os
from dotenv import load_dotenv
```

```python
# Set up API credentials

# Read the Mapbox API key
load_dotenv()
map_box_api = os.getenv("mapbox")

# Set the Mapbox API
px.set_mapbox_access_token(map_box_api)
```

```python
# Read csv data
unicorns_df = pd.read_csv(Path("Resources/CB-Insights_Global-Unicorn-Club_2021.csv"))
locations_df = pd.read_csv(Path("Resources/world_country_and_usa_states_latitude_and_longitude_values.csv")).drop_duplicates()
```

```python
unicorns_df.head()
```

```python
## Prep Geo location data and keep the colums needed for merge
data_to_plot = locations_df[["latitude", "longitude", "Country"]]
data_to_plot
```

```python
#Merging boh datasets to combine latitude and lonigtude for mapbox api to set locaiton
result = pd.merge(data_to_plot, unicorns_df, on=["Country", "Country"])
result
```

```python
# Which country has the most companies
unicorn_country = result["Country"].value_counts()
unicorn_country
```

```python
# Converted into a dataframe and rename the columns
df_unicorn_country = pd.DataFrame(unicorn_country)
df_unicorn_country = df_unicorn_country.reset_index()
df_unicorn_country.columns = ['Country', 'Count']
df_unicorn_country = df_unicorn_country.head(5)
```

```python
# Bar chart showing the count of each country top 5
top_5_fig = px.bar(df_unicorn_country, x='Country', y="Count")
top_5_fig.show()
```

```python
# Map showing all the locations of each company and their size by Billions
map_fig = px.scatter_mapbox(result, lat="latitude", lon="longitude", color="Company", size="Valuation ($B)",
                  color_continuous_scale=px.colors.cyclical.IceFire, size_max=50, zoom=1,
                  mapbox_style="carto-positron")
map_fig.show()
```

```python
# Getting the results for the top 20 valuation funded from each country 
top_countries = result.sort_values(by="Valuation ($B)", ascending=False).head(20)
top_countries
```

```python
# Bar chart shown horitzally top 20 companies valuation from each country
top_20_fig = px.bar(top_countries, x='Valuation ($B)', y='Country', orientation="h")
top_20_fig.show()
```

```python
# Out of 765 private companies which are the top 5 companies by valuation
top_companies = result.sort_values(by="Valuation ($B)", ascending=False).head(5)
top_companies
```

```python
# Bar chart showing the top 5 companies 
top_5_fig = px.bar(top_companies, x='Valuation ($B)', y='Company', orientation="h")
top_5_fig.show()
```

```python
# What is the count of each category
unicorn_category = result["Category"].value_counts()
unicorn_category
```

```python
# Converted into a dataframe and rename the columns
df_unicorn_category = pd.DataFrame(unicorn_category)
df_unicorn_category = df_unicorn_category.reset_index()
df_unicorn_category.columns = ['Category', 'Count']
df_unicorn_category
```

```python
# Pie Chart showing the percentage of each Unicorn Category
percentage_fig = px.pie(df_unicorn_category, values="Count", names="Category", title="Top Category ", width=800, height=400)
percentage_fig.show()
```

```python
# Bar chart showing each count of each category
category_fig = px.bar(df_unicorn_category, x='Category', y="Count")
category_fig.show()
```
